```python
from Problem_1.logparser import LogParser
parser = LogParser()
```

```python
data = []
data.append(parser.parse_log("./Logs/baseline_BdJPsiKs_MagD.log", "PrChecker.Downs"))
data.append(parser.parse_log("./Logs/DR_BdJPsiKs_MagD_1k.log", "PrChecker.Downs"))
data.append(parser.parse_log("./Logs/DR_NE_BdJPsiKs_MagD.log", "PrChecker.Downs"))
data.append(parser.parse_log("./Logs/DR_NE_BdJPsiKs_MagD_1k.log", "PrChecker.Downs"))
```

```python
%matplotlib inline
import matplotlib.pyplot as plot
ax = plot.subplot(111)
colors = ["#ff0000","#ff4e00","#ff9c00","#ffeb00","#c4ff00","#75ff00","#27ff00","#00ff27","#00ff75","#00ffc4","#00ebff","#009cff","#004eff","#0000ff"]
for data in data:
    for i in range(14):
        plot.scatter(float(data[i][2]), float(data[i][1]), s = 25, c = colors[i], label = data[i][0])
    plot.xlabel("Hit efficiency [%]")
    plot.ylabel("Purity [%]")
    plot.legend(loc = 'lower right', scatterpoints = 1, prop={'size':6})
    plot.show()
    plot.clf()
```
